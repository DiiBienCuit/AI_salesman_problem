```python
from itertools import permutations
import pandas as pd 
import numpy as np
import random
import math

def perm(a, k=0):
    if k == len(a):
        print(a)
    else:
        for i in xrange(k, len(a)):
            a[k], a[i] = a[i] ,a[k]
            perm(a, k+1)
            a[k], a[i] = a[i], a[k]
            
def getAllRoutes(city,cities,cityDebut):
    list1 = list(permutations(cities))
    fac = math.factorial(city) - 1
    for x in range(fac,-1,-1):
        if list1[x][0] != cityDebut:
            del list1[x]
        else:
            list1[x] = list1[x] + (cityDebut,)
    return list1

class route:
    def __init__(self,matrix,routeArray):
        self.routeArray = routeArray
        self.matrix = matrix
        
    def calculateRoute(self):
        distance = 0
        for x in range(len(self.routeArray)-1):
            element = self.routeArray[x]-1
            element2 = self.routeArray[x+1]-1
            distance = distance + self.matrix.iloc[element,element2]
        
        return distance
```
